```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
x = np.linspace(-10, 10, 1000)
y_old = np.exp(x)
y_new = 1. + np.exp(x)
```

```python
%matplotlib inline
```

```python
plt.plot(x, y_old, label="$\exp(x)$")
plt.plot(x, y_new, label="$1+\exp(x)$")
plt.minorticks_on()
ylim_l = 0.5
plt.ylim(ylim_l, 2.25)
plt.xlim(-1, 10)
plt.yticks(np.arange(ylim_l, 2.25, 0.125))
plt.grid(which='major')
plt.grid(which='minor', linestyle=':', linewidth='0.5', color='gray')
plt.legend()
plt.show()
```
